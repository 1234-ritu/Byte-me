```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
df=pd.read_csv("student info2.csv")
df
```

```python
df=df.drop(['Timestamp','Name','Gender','Department',"Year","How much time does it take for you to reach college?","What is your preferred mode of transportation to reach college?","If yes, what is your internet speed?","Internet availability at home","How much would you consider as your average attendance throughout all the semesters so far?","Teacher's Feedback","How socially active are you?"],axis=1)
```

```python
df=df.drop(["How many hours per week do you spend on writing assignments?","How many hours per week do you spend on studies?"],axis=1)
```

```python
df=df.drop(["Can you sit a lecture for 2 hrs straight?","Do you do your submissions on time?","If there are 5 hrs of lectures per day, would you attend all?","If there are 5 hrs of practicals per day, would you attend all?","Have you enrolled for coaching classes?","Have you won or participated in any Non - Technical event?"],axis=1)
```

```python
df=df.fillna(0)
df
```

```python
X=df[df.columns[0:15]]
y=df["Label"]
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.4, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
sc=regressor.score(X_test,y_test)
sc1=regressor.score(X_train,y_train)
print("Train Accuracy",sc1*100,"%")
print("Test Accuracy:",sc*100,"%")
```

```python
y_pred = regressor.predict(X_test)
dd = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
dd 
```
